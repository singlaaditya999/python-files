```python
import bs4 
import requests

my_url='https://www.ndtv.com/business/rbi-raises-near-term-inflation-forecast-to-3-4-for-second-quarter-of-2019-20-2111869'

uClient=requests.get(my_url)
print(uClient.text)
page_soup=bs4.BeautifulSoup(uClient.text,"html.parser")

containers=page_soup.findAll("div",{"class":"newcont2"})

container=containers[0]
print(container.div["ins_headline"])
headLine=container.findAll("div",{"class":"ins_descp"})
print(headLine.text)
```
