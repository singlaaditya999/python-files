```python
import numpy as np

X = np.random.rand(40,5)*2 - 10
Y = np.random.rand(40,1)*3 - 10
print(Y.shape)


def sig(z):
    return 1/(1+np.exp(-z))

def derivative_sigmoid(a):
    c=sig(a)*(1-sig(a))
    return c
```

```python
lO_1 = 10 #no of nodes in layer 1
lO_2 = 1 #outpput layer
w_1 = np.random.rand(lO_1,len(X[0])) # layer 1 weight
b_1 = np.random.rand(lO_1) # layer 1 bias
w_2 = np.random.rand(lO_2,lO_1) # layer 2 weight
b_2 = np.random.rand(lO_2) # layer 1 bias
```

```python
def forward_pass(X,lO_1,lO_2,w_1,w_2,b_2,b_1,lr):
    error =np.eye(len(X[0]))
    for i in range(len(X[0])):
        
        a_1 = np.dot(w_1,X[0]) + b_1

        out_1 = sig(a_1)

        a_2 = np.dot(w_2,out_1) + b_2
        out_2 = sig(a_2)

        error = (Y[i] - out_2)**2
#     print(w_1,w_2,b_1,b_2)
    error = np.sum(error)/len(X[0])
    print(error)
    return error,a_1,a_2,out_1,out_2
```

```python
def back_prop(w_1,w_2,b_1,b_2,error,a_1,a_2,out_1,out_2,lr):
    dw_2 = derivative_sigmoid(a_2)*out_1 # change in w_1
    db_2 = derivative_sigmoid(a_2)# change in b_1
    dw_1 = derivative_sigmoid(a_2)*derivative_sigmoid(a_1).reshape(lO_1,1)*X[0].reshape(1,len(X[0]))# change in w_2
    db_1 = derivative_sigmoid(a_2)*derivative_sigmoid(a_1)# change in b_2
    w_1 = w_1 -lr*dw_1 # weight 1 update
    b_1 = b_1 -lr*db_1# bias 1 update
    w_2 = w_2 -lr*dw_2# weight 2 update
    b_2 = b_2 -lr*db_2# bias 2 update
    return w_1,w_2,b_1,b_2
```

```python
np.random.seed(555)
lr =1000000000000
for i in range(1000):
    error,a_1,a_2,out_1,out_2 = forward_pass(X,lO_1,lO_2,w_1,w_2,b_2,b_1,lr)
    w_1 , w_2, b_1 , b_2  = back_prop(w_1,w_2,b_1,b_2,error,a_1,a_2,out_1,out_2,lr)
```
