```python
class aman:
    def __init__(self,w):
        self.w=w
        
    def reval(self,arr):
        l=[]
        row=len(arr)
        column=len(arr[0])
        for i in range(0,row):
            for j in range(0,column):
                l.append(arr[i][j])
            
        return l
    def flat(self,n):
        ar=aman.reval(self,self.w)
        return ar[n]
        
        
    def reshape(self,arr,row,column):
        
        l=[]
        ar= aman.reval(self,arr)
        if len(ar)==(row*column):
            d=0
            for i in range(0,row):
                k=[]
                for j in range(0,column):
                    k.append(ar[d])
                    d+=1

                l.append(k)
            return(l)
        else:
            print("Invalid Shape")
            
    def transpose(self,arr):
        l=[]
        ro=len(arr)
        try:
            colum=len(arr[0])
        except :
            colum=1
        if colum==1:
    
                for j in range(0,ro):
                    k=[]
                    k.append(arr[j])
                    l.append(k)
                return l
        else:    
            for i in range(0,colum):
                for j in range(0,ro):
                    l.append(arr[j][i])
        li=[]
        if len(l)==(ro*colum):
            d=0
            for i in range(0,colum):
                k=[]
                for j in range(0,ro):
                    k.append(l[d])
                    d+=1

                li.append(k)
    
        return li
    
    def stack(self,arr1,arr2,axis):
        axis=str(axis)
        l=[arr1,arr2]
        if axis == "0":
            return l
        elif axis == "-1":
            return aman.transpose(self,l)
        else:
            return "Invalid"
    
    def zero(self,row,column):
        l=[]
        for i in range(0,row):
            k=[]
            for j in range(0,column):
                k.append(0)
            l.append(k)    
        return l
    
    def ones(self,row,column):
        l=[]
        for i in range(0,row):
            k=[]
            for j in range(0,column):
                k.append(1)
            l.append(k)    
        return l
    
    def identiy(self,row):
        l=[]
        for i in range(0,row):
            k=[]
            for j in range(0,row):
                if i==j:
                    k.append(1)
                else:
                    k.append(0)
            l.append(k)    
        return l
    
    def multiply(self,arr1,arr2):
        result=[]
        
        if len(arr1[0])==len(arr2):
            result=aman.zero(self,len(arr1),len(arr2[0]))
            for i in range(len(arr1)):
               
               for j in range(len(arr2[0])):
                   
                   for k in range(len(arr2)):
                        result[i][j] += arr1[i][k] * arr2[k][j]
        else:
            print("Order is Not same")
        for r in result:
            print(r)
        
        
        
            
        
            


def main():
    a=[[0, 1, 2], [3, 4, 5], [5, 7, 8], [9, 10, 11]]
    a1=[[0, 1, 2,8], [3, 4, 5,7], [5, 7, 8,8]]
    
    a2=[0, 3, 5]
    d=aman(a)
    re=d.reshape(a,3,4)
    l=d.transpose(a2)
    d.multiply(a,a1)
    lo=d.identiy(4) 
    for i in lo:
        print(i)
if __name__ =="__main__":
    main()
```
