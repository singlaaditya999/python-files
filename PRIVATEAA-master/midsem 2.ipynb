```python
def trace(list1):
    if len(list1)==len(list1[0]):
        print('Trace is:')
        for i in range (0,len(list1)):
            print(list1[i][i],',')
            
def isIdentity(list1):
    minimum=min(len(list1),len(list1[0]))
    for i in range (0,len(list1)):
        for j in range (0,list1[0]):
            if list1[i][j]!=1 && i==j:
                print('Not Identity')
            if list1[i][j]!=0:
                print('Not Identity')
    print('Is identity')
def shape(arr):
        rows=len(arr)
        column=len(arr[0])
        print('No. of rows=',rows)
        print('No. of colums=',column)
        
def makeMatrix(rows,columns):
    list1=[0]
    
    list2=[]
    for j in range(0,columns):
            list2=list2+list1
    listFinal=[]
    for i in range(0,rows):
            listFinal=listFinal+[list2]
    
    for i in range(0,rows):
        for j in range(0,columns):
                print('Enter data of[',i,j,']')
                data=int(input())
                listFinal[i][j]=data
    return listFinal
            
def addition(arr1,arr2):
    if len(arr1)!=len(arr2):
        return []
    if len(arr1[0])!=len(arr2):
        return []
    list1=[0]
    rows=len(arr1)
    columns=len(arr1[0])
    list2=[]
    for j in range(0,columns):
            list2=list2+list1
    listFinal=[]
    for i in range(0,rows):
            listFinal=listFinal+[list2]
    for i in range(0,rows):
        for j in range(0,columns):
            listFinal[i][j]=arr1[i][j]+arr2[i][j]
    return listFinal

def subtraction(arr1,arr2):
    if len(arr1)!=len(arr2):
        return []
    if len(arr1[0])!=len(arr2):
        return []
    list1=[0]
    rows=len(arr1)
    columns=len(arr1[0])
    list2=[]
    for j in range(0,columns):
            list2=list2+list1
    listFinal=[]
    for i in range(0,rows):
            listFinal=listFinal+[list2]
    for i in range(0,rows):
        for j in range(0,columns):
            listFinal[i][j]=arr1[i][j]-arr2[i][j]
    return listFinal

def printMatrix(list1):
    if(list1==[]):
        print('matrix is empty')
    for i in range(0,len(list1)):
        for j in range(0,len(list1[0])):
            print(list1[i][j],',')
        print('\n')

def main():
    columns=int(input('Enter number of columns'))
    rows=int(input('Enter number of rows'))
    matrix1=makeMatrix(rows,columns)
    shape(matrix1)
    columns2=int(input('Enter number of columns'))
    rows2=int(input('Enter number of rows'))
    matrix2=makeMatrix(rows2,columns2)
    added=addition(matrix1,matrix2)
    printMatrix(added)
    trace(matrix1)
    #a1=[[0, 1, 2,8], [3, 4, 5,7], [5, 7, 8,8]]
    #print(shape(d))
    
    
if __name__ =="__main__":
    main()
            


```
